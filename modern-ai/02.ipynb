```python
from torchvision import datasets
```

```python
mnist_train = datasets.MNIST(root="data", train=True, download=True)
mnist_test = datasets.MNIST(root="data", train=False, download=True)
```

```python
X_train = ((mnist_train.data)/255.).reshape(60000, 784)
y_train = mnist_train.targets
X_test = ((mnist_test.data)/255.).reshape(10000, 784)
y_test = mnist_test.targets
```

```python
W = torch.randn(10, 784)
for i in range(10):
    W[i] = X_train[y_train == i].mean(dim=0)
    W[i] = W[i] / W[i].norm()
```

```python
def zero_one_loss(y_hat, y):
    return (y_hat.argmax(dim=1) != y).float().mean()

def cross_entropy_loss(y_hat, y):
    loss = 0.0
    for i in range(len(y)):
        loss += -y_hat[i, y[i]] + y_hat[i].exp().sum().log()
    return loss / len(y)
```

```python
W_zero = torch.zeros(10, 784)
print(cross_entropy_loss(X_test @ W_zero.T, y_test))
print(cross_entropy_loss(X_test @ W.T, y_test))
```
