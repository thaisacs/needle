```python
class Linear(Module):
    def __init__(self, in_dim, out_dim):
        super().__init__()
        self.weight = Parameter(torch.randn(out_dim, in_dim) * math.sqrt(2/in_dim))
    
    def forward(self, X):
        return X @ self.weight.T
```

```python
X = torch.randn(100, 128)
layer = Linear(128, 256)
layer(X).shape
```
